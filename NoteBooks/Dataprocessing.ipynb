```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
data_path=Path("../synthetic_food_dataset_imbalanced.csv")
raw_data=pd.read_csv(data_path)
```

```python
missing_counts = raw_data.isna().groupby(raw_data['Food_Name']).sum()
print(missing_counts)
```

```python
path=Path('../Data/synthetic_food_dataset_imbalanced.csv')
df = pd.read_csv(path)
df_cleaned = df.drop_duplicates()
numeric_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    df_cleaned[col] = df_cleaned[col].fillna(df_cleaned.groupby('Food_Name')[col].transform('median'))
# df_cleaned[numeric_cols] = df_cleaned[numeric_cols].fillna(df_cleaned[numeric_cols].median())

df_cleaned.to_csv('../Data/cleaned_food_data.csv', index=False)
print("\nCleaned data has been saved to 'cleaned_food_data.csv'")
```
