```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
data_path=Path("../Data/synthetic_food_dataset_imbalanced.csv")
raw_data=pd.read_csv(data_path)
```

```python
missing_counts = raw_data.isna().groupby(raw_data['Food_Name']).sum()
print(missing_counts)
```

```python
numerical_features = [
        'Calories', 'Protein', 'Fat', 'Carbs', 'Sugar', 'Fiber', 
        'Sodium', 'Cholesterol', 'Glycemic_Index', 'Water_Content', 'Serving_Size'
    ]
# numeric_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
categorical_features = ['Meal_Type', 'Preparation_Method']
boolean_features = ['Is_Vegan', 'Is_Gluten_Free']
target_feature = 'Food_Name'
```

```python
print((raw_data[numerical_features] < 0).sum())
```

```python
df_cleaned = raw_data.drop_duplicates()
```

```python
for col in numerical_features:
    df_cleaned[col] = df_cleaned[col].fillna(df_cleaned.groupby('Food_Name')[col].transform('median'))
```

```python
print((df_cleaned[numerical_features] < 0).sum())
```

```python
df_cleaned.head()
```

```python
missing_counts = df_cleaned.isna().groupby(raw_data['Food_Name']).sum()
print(missing_counts)
```

```python
boundries={}
for col in numerical_features:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = max(0, Q1 - 1.5 * IQR)
    upper_bound = Q3 + 1.5 * IQR
    print(f"Processing column: {col}")
    print(lower_bound)
    print(upper_bound)
    boundries[col]=(lower_bound,upper_bound)
    # Clip the data using the IQR bounds
    df_cleaned[col] = df_cleaned[col].clip(lower=lower_bound, upper=upper_bound)
print(boundries)
print("Outlier clipping complete using IQR method.")
```

```python
output_dir = Path('../Data/DataProcessing')
output_dir.mkdir(exist_ok=True)
df_cleaned.to_csv('../Data/DataProcessing/cleaned_food_data.csv', index=False)
```

```python
import pickle
metadata={
    'numerical_features': numerical_features,
    'categorical_features': categorical_features,
    'boolean_features': boolean_features,
    'target_feature': target_feature,
    'outlier_boundaries':boundries
}
with open('../Data/DataProcessing/metadata.pkl', 'wb') as f:
    pickle.dump(metadata, f)
```
