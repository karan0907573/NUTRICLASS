```python
models = [{'trained_model':logistic_regression, 'name': 'Logistic Regression'},{'trained_model':decision_tree, 'name': 'Decision Tree'},{'trained_model':random_forest, 'name': 'Random Forest'},{'trained_model':xgboost, 'name': 'XGBoost'},{'trained_model':gradient_boosting, 'name': 'Gradient Boosting'},{'trained_model':knn_model, 'name': 'KNN'},{'trained_model':svm_model, 'name': 'SVM'}]
```

```python
import pandas as pd
```

```python
results_list = []

for model in models:
    results_list.append({
        'Model': model['name'], 
        'Test Accuracy': model['trained_model']['results']['test_accuracy']
    })
    print(model['name'], model['trained_model']['results']['test_accuracy'])
accuracy_df = pd.DataFrame(results_list)
accuracy_df = accuracy_df.sort_values(by='Test Accuracy', ascending=False)
accuracy_df = accuracy_df.reset_index(drop=True)

print(accuracy_df)
```
